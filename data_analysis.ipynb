```python
import pandas as pd
```

```python
# Load datasets
customers_df = pd.read_csv("Customers.csv")
orders_df = pd.read_csv("Orders.csv")
products_df = pd.read_csv("Products.csv")
```

```python
# Total Orders by Customer : How many orders has each customer placed?
orders_df.groupby("CustomerID")["OrderID"].nunique()
```

```python
# 2.	Average Quantity per Product
# What is the average quantity ordered for each product?
orders_df.groupby("ProductID")["Quantity"].mean()
```

```python
# Q1: Total revenue per customer & top customer
customer_revenue = orders_df.groupby("CustomerID")["TotalSales"].sum().reset_index()
display(customer_revenue)
top_customer = customer_revenue.sort_values(by="TotalSales", ascending=False).head(1)
display(top_customer)
top_customer_info = top_customer.merge(customers_df, on="CustomerID", how="left")
top_customer_info
```

```python
# Q2: Repeat customers
repeat_customers = orders_df.groupby("CustomerID")["OrderID"].nunique().reset_index()
display(repeat_customers)
repeat_customers = repeat_customers[repeat_customers["OrderID"] > 1]
display(repeat_customers)
repeat_percentage = round((len(repeat_customers) / customers_df["CustomerID"].nunique()) * 100, 2)
display(repeat_percentage)
```

```python
# Q4: Top 5 products by quantity
product_sales = orders_df.groupby("ProductID")["Quantity"].sum().reset_index()
top_products = product_sales.sort_values(by="Quantity", ascending=False).head(5)
top_products = top_products.merge(products_df, on="ProductID", how="left")
top_products
```
