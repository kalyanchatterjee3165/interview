Q2. You are given a word (string).
Write a Python function that uses the enumerate() function to find the index of the first vowel (a, e, i, o, u) in the word.
If no vowel is found, return -1.

def find_first_vowel_index(word):
    pass

Test Case: 
find_first_vowel_index("apple")     âž 0  
find_first_vowel_index("strength")  âž 3  
find_first_vowel_index("rhythm")    âž -1  

```python
def find_first_vowel_index(word):
    vowels = 'aeiou'
    for index, char in enumerate(word.lower()):
        if char in vowels:
            return index
    return -1

# Test Cases
print(find_first_vowel_index("apple"))     # Output: 0
print(find_first_vowel_index("strength"))  # Output: 3
print(find_first_vowel_index("rhythm"))    # Output: -1
```

Q2: 

You are given a string s consisting of only lowercase letters. Your task is to process this string based on the following rules:
	1.	Remove all repeating characters, keeping only the first occurrence of each character.
	2.	Extract the first three unique characters from the resulting string.
	3.	If the number of unique characters is less than 3, pad the result with underscores ('_') at the end to make it exactly 3 characters long.

Implement the following function:
def extract_unique_three(s: str) -> str:
    pass


Follow up :
1. Create a Dataframe out of this data
df_data = {
    "Company": ["Amazon", "Google", "Facebook", "Netflix", "LG"],
    "Revenue_2024": [574.8, 891.2, 116.3, 40.2, 394.3]
}

2. add a new column : "Company_Code" by using this funtion that you have just created

```python
import pandas as pd

# Step 1: Function definition
def extract_unique_three(s: str) -> str:
    seen = set()
    unique_chars = []
    
    for char in s:
        if char not in seen:
            seen.add(char)
            unique_chars.append(char)
    
    result = ''.join(unique_chars[:3])
    return result.ljust(3, '_')

# Step 2: Create sample company data
df_data = {
    "Company": ["Amazon", "Google", "Facebook", "Netflix", "LG"],
    "Revenue_2024": [574.8, 891.2, 116.3, 40.2, 394.3]
}

# Create DataFrame
df = pd.DataFrame(df_data)

# Add 'Code' column using the function
df["Code"] = df["Company"].str.lower().apply(extract_unique_three)

# Display final DataFrame
print(df)
```

Q3. 
Write a function that takes a block of text (string) as input and returns the most frequently occurring word.
In case of a tie (multiple words with the same highest frequency), return the alphabetically first word.

You should:
	1.Normalize the text: convert to lowercase.
	2.Remove punctuation.
	3.Tokenize (split) into words.
	4.Count word frequencies.
	5.Return the most frequent word (alphabetically smallest in case of tie).


def most_frequent_word(text: str) -> str:
    pass


Test Cases: 

print(most_frequent_word("The cat and the hat"))                           # âž "the"
print(most_frequent_word("Hello hello world!"))                            # âž "hello"
print(most_frequent_word("One fish two fish red fish blue fish"))         # âž "fish"
print(most_frequent_word("To be, or not to be: that is the question."))   # âž "be"
print(most_frequent_word("Data science is fun. Science is also challenging."))  # âž "science"
print(most_frequent_word("apple ball apple ball cat"))                    # âž "apple" (tie: apple, ball â†’ apple is alphabetically first)

â¸»


```python
import re
from collections import Counter

def most_frequent_word(text: str) -> str:
    # Normalize text: lowercase + remove punctuation
    cleaned = re.sub(r'[^\w\s]', '', text.lower())
    words = cleaned.split()
    
    # Count word frequencies
    word_counts = Counter(words)
    
    # Find max frequency
    max_freq = max(word_counts.values(), default=0)
    
    # Get all words with max frequency
    most_frequent_words = [word for word, count in word_counts.items() if count == max_freq]
    
    # Return the alphabetically first word among the most frequent
    return min(most_frequent_words)
```

```python

orders_per_customer = orders_df.groupby("CustomerID")["OrderID"].nunique().reset_index()
orders_per_customer.columns = ["CustomerID", "TotalOrders"]
orders_per_customer
```

```python

# Merge with customer names
orders_per_customer = orders_per_customer.merge(customers_df[["CustomerID", "CustomerName"]], on="CustomerID", how="left")
orders_per_customer
```

```python
# Q2 Identify customers who have placed more than one distinct order. Calculate and report the percentage of such repeat customers out of the total customer base.
repeating_customers = orders_per_customer[orders_per_customer["TotalOrders"] > 1]

# Calculate percentage
total_customers = customers_df["CustomerID"].nunique()
repeat_percentage = round((len(repeating_customers) / total_customers) * 100, 2)

print("Repeating Customers:")
display(repeating_customers[["CustomerID", "CustomerName", "TotalOrders"]])

print(f"\n% of Repeating Customers: {repeat_percentage}%")
```

```python

# Q1. Can you explain the code
import time
from functools import wraps

def timing(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} took {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timing
def slow_function():
    time.sleep(1)
    return "Done"
```

```python
# Q2. Do you know generator Funtion?  Print fibinacci series using Generator Funtion

def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci_generator(10):
    print(num, end=' ')
```
